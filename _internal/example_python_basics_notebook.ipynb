```python
import numpy as np
import ndv
import matplotlib.pyplot as plt
```

```python
img = np.random.randint(0, 255, (10, 100, 100), dtype=np.uint8)
print(img.shape, img.dtype)
```

```python
ndv.imshow(img)
```

```python
viewer = ndv.imshow(img)
```

```python
viewer.widget().children[1].snapshot()
```

```python
plt.imshow(img[0], cmap="gray")
plt.axis("off")
plt.show()
```
